```python
import numpy as no
from sklearn.linear_model import Perceptron
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
iris_data
```

```python
X = iris_data.data
y = iris_data.target
#y = y == 0
y
```

```python
perceptron = Perceptron()
perceptron.fit(X,y)
```

```python
perceptron.score(X,y)
```
